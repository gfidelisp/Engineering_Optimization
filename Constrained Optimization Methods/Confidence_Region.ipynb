# Optimization Problem 

Minimize the cost of operation of the 3 power plants, given a power requirement L

### Objective function

```python
def f(C,P, L):
         
        func = ((C[0][0]*P[0] + C[0][1]*P[0]**2 + C[0][2]*P[0]**3 + C[0][3]*P[0]**4) +
                (C[1][0]*P[1] + C[1][1]*P[1]**2 + C[1][2]*P[1]**3 + C[1][3]*P[1]**4) +
                (C[2][0]*P[2] + C[2][1]*P[2]**2 + C[2][2]*P[2]**3 + C[2][3]*P[2]**4) -
                (P[0] + P[1] + P[2] - L) + 0.125*(P[0] + P[1] + P[2] - L)**2  
        )

        return func
```

### Gradient of the objective function

```python
def g(C, P, L):
    g1 = ((C[0][0] + 2*C[0][1]*P[0] + 3*C[0][2]*P[0]**2 + 4*C[0][3]*P[0]**3) - 1 + 0.25*(P[0] + P[1] + P[2] - L))
    g2 = ((C[1][0] + 2*C[1][1]*P[1] + 3*C[1][2]*P[1]**2 + 4*C[1][3]*P[1]**3) - 1 + 0.25*(P[0] + P[1] + P[2] - L))
    g3 = ((C[2][0] + 2*C[2][1]*P[2] + 3*C[2][2]*P[2]**2 + 4*C[2][3]*P[2]**3) - 1 + 0.25*(P[0] + P[1] + P[2] - L))

    return np.array([g1, g2, g3])
```

#### Hessian of the objective function

```python
def H(C,P,L):

    h1 = ((2*C[0][1] + 6*C[0][2]*P[0] + 12*C[0][3]*P[0]**2) + 0.25)
    h2 = ((2*C[1][1] + 6*C[1][2]*P[1] + 12*C[1][3]*P[1]**2) + 0.25)
    h3 = ((2*C[2][1] + 6*C[2][2]*P[2] + 12*C[2][3]*P[2]**2) + 0.25)

    Hes = [
        [h1  , 0.25, 0.25],
        [0.25, h2  , 0.25],
        [0.25, 0.25, h3  ]
    ]

    return Hes
```

## Confidence region

```python
def T(Tau, Data):
    p_u, p_b, Delta = Data  
    vec = p_u + (Tau-1)*(p_b-p_u)
    return(np.linalg.norm(vec)**2 - Delta**2)

def m(f0, g0, B0, p0):
    return f0 + g0@p0.reshape(-1,1) + p0@B0@p0.reshape(-1,1)

def m_scipy(p0, Data):
    f_0, g_0, B_0 = Data  
    return f_0 + g_0@p0.reshape(-1,1) + p0@B_0@p0.reshape(-1,1)

def constraint(p0):
    return Delta - np.linalg.norm(p0) 
```

```python
# Power requirement
L = 25

# Maximum radius
Delta_max = 10

# Maximum iteration
it_max = 100

#Initial guess
X_0 = [L/3, L/3, L/3]
it = 0
X = X_0
global Delta

#Empty arrays
X_array =  []
f_array = []
g_array = []
Delta_array = []
rho_array = []

# Methods
met = 'DogLeg'
Delta = 0.5
eta = 0

g_0 = g(C, X, L)
p  =[1,1,1]
X_array.append(X)
Delta_array.append(Delta)

while it < it_max and np.linalg.norm(g_0) > 10**-5 and np.linalg.norm(p) > 1.e-8:
    f_0 = f(C, X, L)
    g_0 = g(C, X, L)
    B_0 = H(C, X, L)

    f_array.append(f_0)
    g_array.append(np.linalg.norm(g_0))

    if met == 'DogLeg':
        p_b = - np.dot(np.linalg.norm(B_0),g_0) 
        p_u = - np.dot(g_0,g_0)/np.dot(np.dot(g_0, B_0),g_0)*g_0 

        if np.linalg.norm(p_b) <= Delta:
            p = p_b
    
        else:
            tau = fsolve(T,args = np.array([p_u, p_b, Delta], dtype='object'), x0 = [1.5], maxfev = 10000000)
            p = p_u + (tau-1)*(p_b-p_u)

    elif met == 'Exact':
        p = minimize(m_scipy,  args = [f_0, g_0, B_0 ],  x0 = [0, 0, 0], method = "SLSQP",
                     constraints={"fun": constraint, "type": "ineq"}).x

    rho = (f_0 - f(C, X + p, L))/(f_0 - m(f_0, g_0, B_0, p))
    

    if rho < 0.25:
        Delta = 0.25*Delta
        
    elif rho > 0.75 and isclose(np.linalg.norm(p), Delta, rel_tol =1e-1):
        Delta = np.minimum(2*Delta, Delta_max)

    

    if rho > eta:
        X = X + p
    
    X_array.append(X)
    Delta_array.append(Delta)
    rho_array.append(rho)

    it =  it + 1
```
