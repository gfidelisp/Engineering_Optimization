###  "Banana" function

```python
def fban1(*x):
    f = 10*x[0]**4 - 20*x[0]**2*x[1] + 10*x[1]**2 + x[0]**2 -2*x[0] + 5
    return f
```

### Gradient of the "Banana" function

```python
def gban1(*x):
    
    gf = np.zeros(2)
    gf[0] = 40*x[0]**3 - 40*x[0]*x[1]+2*x[0]-2
    gf[1] = -20*x[0]**2+20*x[1]
    
    return gf.T
```

### Hessian of the "Banana" function

```python
def hban1(*x):
    hf = np.zeros([2,2])
    hf[0,0] = 120*x[0]**2 - 40*x[1] +2 
    hf[1,1] = 20 
    hf[0,1] = -40*x[0]
    hf[1,0] = -40*x[0]
    
    return hf
```

#### Line search by the Golden Search

```python
def GldSrc(Fnome, X, d):
    i = 0
    sigma  = 1
    al = 0
    au = sigma  
    while Fnome(*(X + au*d)) < Fnome(*(X + al*d)):
        i = i+1
        al = au
        au = au + 1.618**(i)*sigma
    if i > 0:
        al = au - sigma*(1.618**i + 1.618**(i-1))

    error = 1
    tresh = 1.e-3
    while error > tresh:
        i = i+1
        aa = al + 0.382*(au-al)
        ab = al + 0.618*(au-al)

        Xa = X + aa*d
        Xb = X + ab*d
        if Fnome(*Xa) < Fnome(*Xb):
            au = ab
        elif Fnome(*Xb) <= Fnome(*Xa):
            al = aa
        error = np.abs(au - al)

    return au, X + au*d, i 
```

####  Line search by Quadratic Interpolation

```python
def LSPolynomial(Fnome, X, d):
    i = 0
    sigma  = 1
    al = 0
    au = sigma
    
    while Fnome(*(X + au*d)) < Fnome(*(X + al*d)):
        i = i+1
        al = au
        au = au + 1.618**(i)*sigma
    
    if i > 0:
        al = au - sigma*(1.618**i + 1.618**(i-1))

    error = 1
    tresh = 1.e-3

    while error > tresh:
        i=i+1
        ai = al + 0.5*(au-al)
    
        f_Xl = Fnome(*(X + d*al))
        f_Xi = Fnome(*(X + d*ai))
        f_Xu = Fnome(*(X + d*au))

        a_2 = 1/(au - ai)*( ((f_Xu - f_Xl) / (au - al)) - ( (f_Xi - f_Xl)/(ai - al) ) )
        a_1 = ((f_Xi - f_Xl)/(ai - al)) - a_2*(al+ai)

        if a_2 > 0:
            a = -a_1/(2*a_2)
        else:
            break
        
        if ai < a:
            if f_Xi <= Fnome(*(X + d*a)):
                au = a
            elif f_Xi > Fnome(*(X + d*a)):
                al = ai
        elif ai > a:
            if f_Xi <= Fnome(*(X + d*a)):
                al = a
            elif f_Xi > Fnome(*(X+d*a)):
                au = ai
                
        error = np.abs(au - al)
        
    return au, X + d*au, i 
```

####  Line search by Armijo Rule

```python
def LSArmijo(Fnome, X, d,g):
    rho1 = 0.9
    rho2 = 0.1
    nUP = 2          # coefficient for interval increment
    ndown = 0.7      # coefficient for backtracking
    
    g0 = g           # f'(0)
    f0 = Fnome(*X)    # f(0)
    s = 1            # Initial Step value
    
    Xs  = X + s*d     # X(s)
    fXs = Fnome(*Xs)   # f(Xs) 
    
    #DENTRO DA condição (9.4)livro do ARORA
    data = []
    gfd = np.dot(np.array(g0).T,d)
    i = 0
    #if gfd > 0:
        #print('No descent direction')
    data.append([s,fXs,f0 + s*rho1*np.dot(np.array(g0),d),f0 + s*rho2*np.dot(np.array(g0),d)])
    while fXs <= f0 + s*rho1*np.dot(np.array(g0),d):
        
        s = nUP*s
        Xs = X + s*d
        fXs = Fnome(*Xs)
        i=i+1
        data.append([s,fXs,f0 + s*rho1*np.dot(np.array(g0),d),f0 + s*rho2*np.dot(np.array(g0),d)])
    
    while fXs > f0 + s*rho2*np.dot(np.array(g0),d):
        
        s = ndown*s
        Xs = X+s*d
        fXs = Fnome(*Xs)
        i=i+1
        data.append([s,fXs,f0 + s*rho1*np.dot(np.array(g0),d),f0 + s*rho2*np.dot(np.array(g0),d)])
    return s, Xs, i
```

## Descent direction search by Gradient and Newton's method

```python
def Newton(Fnome, Gnome, Hnome, X0, Type,LS,Silent = False):
    
    # Maximum iteration number
    NIter = 1500;

    # Gradient of cost function at X0
    g0 = Gnome(*X0)
    
    #Convergence test at X0
    Epsilon  = 1.e-8*np.linalg.norm(g0)
    
    if np.linalg.norm(g0) < Epsilon:
        return 
    
    #Inicialization
    X = X0
    dim = np.size(X)
    H = np.identity(dim)
    g = Gnome(*X)
    count = 0
    points = []
    #Iterations
    it_line = 0
    for i in range(NIter):
        d = -np.dot(np.linalg.inv(H),g)

        #Line Search
        if LS == 0:
            [s,X1,it_l] = LSArmijo(Fnome, X, d, g)
        elif LS == 1:
            [s,X1,it_l] = GldSrc(Fnome, X, d)
        elif LS == 2:
            [s,X1,it_l] = LSPolynomial(Fnome, X, d)
        
        it_line = it_line + it_l
        
        #Convergence Verification
        
        f1 = Fnome(*X1)
        g1 = Gnome(*X1)
        norma = (np.linalg.norm(g1)/np.linalg.norm(g0))
        
        
        dx = [X[0], X[0]]
        dy = [X[1], X[1]]
        points.append(X)
        
        #plt.plot(dx,dy,'ro')
        if norma < Epsilon:
            if Silent == False:
                print('Convergence OK!')
                print("Descent Iteration number", i)
                print("Line Search Total Iteration number", it_line)
                print("f  =" + str(f1) + " Step = " + str(s))
                print("X  =" + str(X1))
                print("g/go  = " + str(g1))
        
            return (X1,f1,g1,points)  
        


        if Type == 0:
            #Gradient Descent
            H = np.identity(dim)
        elif Type == 1:
            #Newton
            H = Hnome(*X1)
            
        elif Type == 2:
            #BFGS
            pass
            
        elif Type == 3:
            #DFP
            pass
        
        elif Type == 4:
            #Gradiente Conjugado
            pass

        X = X1
        g = g1
           
    return (X,f1,g1,points)
```

# Incorporation of some conditions in Newton's Method

```python
def NewtonCond(Fnome, Gnome, Hnome, X0, Type,LS,Silent = False):
    
    # Maximum iteration number
    NIter = 1500;

    # Gradient of cost function at X0
    g0 = Gnome(*X0)
    
    #Convergence test at X0
    Epsilon  = 1.e-8*np.linalg.norm(g0)
    
    if np.linalg.norm(g0) < Epsilon:
        return 
    
    #Inicialization
    X = X0
    dim = np.size(X)
    H = np.identity(dim)
    g = Gnome(*X)
    count = 0
    points = []
    it_line = 0
    #Iterations
    
    for i in range(NIter):
        Inv = False
        eps = 0.1
        Theta = 0.1
        Beta = 0.1
        while Inv == False:
            try:
                d = -np.dot(np.linalg.inv(H),g)
                if np.dot(g,d) >= -Theta*np.linalg.norm(g)*np.linalg.norm(d):
                    H = H + eps*np.identity(dim)
                    eps = 2*eps
                else:
                    Inv = True
            except:
                H = H + eps*np.identity(dim)
                eps = 2*eps

        if np.linalg.norm(d) < Beta*np.linalg.norm(g):
            d = d/np.linalg.norm(d)*Beta*np.linalg.norm(g)

        #Line Search
        if LS == 0:
            [s,X1,it_l] = LSArmijo(Fnome, X, d, g)
        elif LS == 1:
            [s,X1,it_l] = GldSrc(Fnome, X, d)
        elif LS == 2:
            [s,X1,it_l] = LSPolynomial(Fnome, X, d)
        
        it_line = it_line + it_l
        
        
        #Convergence Verification
        
        f1 = Fnome(*X1)
        g1 = Gnome(*X1)
        norma = (np.linalg.norm(g1)/np.linalg.norm(g0))
        
        dx = [X[0], X[0]]
        dy = [X[1], X[1]]
        points.append(X)        
        if norma < Epsilon:
            if Silent == False:
                print('Convergence OK!')
                print("Descent Iteration number", i)
                print("Line Search Total Iteration number", it_line)
                print("f  =" + str(f1) + " Step = " + str(s))
                print("X  =" + str(X1))
                print("g/go  = " + str(g1))
        
            return (X1,f1,g1,points) 
        
        if Type == 0:
            #Gradient Descent
            H = np.identity(dim)
        elif Type == 1:
            #Newton
            H = Hnome(*X1)
            
        elif Type == 2:
            #BFGS
            pass
            
        elif Type == 3:
            #DFP
            pass
        
        elif Type == 4:
            #Gradiente Conjugado
            pass 

        X = X1
        g = g1
           
    return (X,f1,g1,points)
```

## Gradient of the Rosembrock Function

```python
def gros(*x):
    
    gf = np.zeros(2)
    gf[0] = 400*x[0]**3 - 400*x[0]*x[1] + 2*x[0] - 2
    gf[1] = 200*x[1] - 200 * x[0]**2
    
    return gf.T
```

## Quase-Newton (BFGS and DFP) and Conjugated Gradient

```python
def QNewton(Fnome, Gnome, Hnome, X0, Type,LS,Silent = False):
    
    # Maximum iteration number
    NIter = 1500;

    # Gradient of cost function at X0
    g0 = Gnome(*X0)
    
    #Convergence test at X0
    Epsilon  = 1.e-8*np.linalg.norm(g0)
    
    if np.linalg.norm(g0) < Epsilon:
        return 
    
    #Inicialization
    X = X0
    dim = np.size(X)
    H = np.identity(dim)
    g = Gnome(*X)
    count = 0
    points = []
    it_line = 0
    #Iterations
    
    for i in range(NIter):
        Inv = False
        eps = 0.1
        Theta = 0.1
        Beta = 0.1
        if Type == 1 or Type == 2:
            while Inv == False:
                try:
                    d = -np.dot(np.linalg.inv(H),g)
                    if np.dot(g,d) >= -Theta*np.linalg.norm(g)*np.linalg.norm(d):
                        H = H + eps*np.identity(dim)
                        eps = 2*eps
                    else:
                        Inv = True
                except:
                    H = H + eps*np.identity(dim)
                    eps = 2*eps
                    
        elif Type == 4:
            if i ==0:
                gx = g
                d  = -np.dot(H,g)
                d_ = d
            else:
                fac = (np.linalg.norm(g)/np.linalg.norm(gx))**2
                d  = -np.dot(H,g) + fac*d_
                d_ = d
                gx = g
            
        elif Type == 3:
            d = -np.dot(H,g)
            if np.dot(g,d) >= -Theta*np.linalg.norm(g)*np.linalg.norm(d):
                        H = H + eps*np.identity(dim)
                        eps = 2*eps

        if np.linalg.norm(d) < Beta*np.linalg.norm(g):
            d = d/np.linalg.norm(d)*Beta*np.linalg.norm(g)

        #Line Search
        if LS == 0:
            [s,X1,it_l] = LSArmijo(Fnome, X, d, g)
        elif LS == 1:
            [s,X1,it_l] = GldSrc(Fnome, X, d)
        elif LS == 2:
            [s,X1,it_l] = LSPolynomial(Fnome, X, d)
        
        it_line = it_line + it_l
        
        
        #Convergence Verification
        
        f1 = Fnome(*X1)
        g1 = Gnome(*X1)
        norma = (np.linalg.norm(g1))
        
        dx = [X[0], X[0]]
        dy = [X[1], X[1]]
        points.append(X)
        
        if norma < Epsilon:
            if Silent == False:
                print('Convergence OK!')
                print("Descent Iteration number", i)
                print("Line Search Total Iteration number", it_line)
                print("f  =" + str(f1) + " Step = " + str(s))
                print("X  =" + str(X1))
                print("g/go  = " + str(g1))
        
            return (X1,f1,g1,points)  
        

        

        if Type == 0:
            #Gradient Descent
            H = np.identity(dim)
        elif Type == 1:
            #Newton
            H = Hnome(*X1)
            
        elif Type == 2:
            #BFGS
            s_k = np.reshape(np.array([s*d]),(dim,1))
            c_k = np.reshape(np.array([g]),(dim,1)) 
            c_k1 = np.reshape(np.array([g1]),(dim,1)) 
            y_k = np.reshape(np.array([g1- g]),(dim,1))
            
            
            #H_1 = ((s_k + H@y_k)@s_k.T + s_k@(s_k-H@y_k).T)/(s_k.T@y_k)
            #H_2 = -((np.array(s_k - H@y_k).T)@y_k*s_k@s_k.T)/(s_k.T@y_k)**2
            
            H1 = y_k@y_k.T/(y_k.T@s_k)
            H2 = -(H@s_k@s_k.T@H)/(s_k.T@H@s_k)
            
            Hk = H + H1 + H2
            H = Hk
            
        elif Type == 3:
            #DFP
            s_k = np.reshape(np.array([s*d]),(dim,1))
            y_k = np.reshape(np.array([g1- g]),(dim,1))
            
            Bk = (s_k@s_k.T)/(y_k.T@s_k)
            #Ck = (-z_k@z_k.T)/(y_k@z_k)
            Ck = -(H@y_k@y_k.T@H)/(y_k.T@H@y_k)
            
            Hk= H +Bk + Ck
            H = Hk
        
        elif Type == 4:
            #Gradiente Conjugado
            pass 

        X = X1
        g = g1
           
    return (X,f1,g1,points)
```

## External Penalization

#### Penalized Rosembrock Function

```python
def fros_pen(*x):
    f = (1 - x[0])**2 + 100*(x[1]-x[0]**2)**2 + rho*(max(0,x[0]-0.5)**2 + max(0,x[1]-1)**2)
    return f
```

#### Gradient of the Penalized Rosembrock function

```python
def gros_pen(*x):
    gf = np.zeros(2)
    gf[0] = 400*x[0]**3 - 400*x[0]*x[1] + 2*x[0] - 2 +2*rho*max(0,x[0]-0.5)
    gf[1] = 200*x[1] - 200 * x[0]**2 +2*rho*max(0,x[1]-1)
    
    return gf.T
```

#### Hessian of the  Penalized Rosembrock function

```python
def hros(*x):
    hf = np.zeros([2,2])
    hf[0,0] = 1200*x[0]**2 - 400*x[1] + 2
    hf[1,1] = 200
    hf[0,1] = -400*x[0]
    hf[1,0] = -400*x[0]
    
    return hf
```

```python
#intial point
x0=np.array([-0.5, 0.5]).T

#Ghraphic: plot the function to be minimized
xmin = -2
xmax = 2
dx = .05
ymin = -1 
ymax = 4
dy = .05

[z,y]= np.meshgrid(np.arange(xmin,xmax+dx,dx), np.arange(ymin,ymax+dy,dy));
```
